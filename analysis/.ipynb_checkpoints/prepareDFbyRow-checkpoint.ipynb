```python
def get_row_vals(row):
    array = np.array(row['presented_pixel_data'], dtype=float)
    mean_array = np.mean(array,axis=(0,2))
    return np.isnan(mean_array)

def get_row_vals_three_stimuli(row):
    array = np.array(row['presented_pixel_data'], dtype=float)
    mean_array = np.mean(array,axis=(0,1,3))
    return np.isnan(mean_array)

E2_files = list(pathlib.Path(path.join('..','experiments','Exp2rows','data','jatos_resultfiles_batch1')).glob('**/*.json'))

E2_group_longdf = pd.DataFrame()

for i,file in enumerate(E2_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    occluded_rows = np.array(mydf.apply(get_row_vals, axis=1).to_list())
    
    n, m = occluded_rows.shape
    longdf = pd.DataFrame(np.repeat(mydf.values, m, axis=0), columns=mydf.columns)
    longdf ['row_index'] = np.tile(np.arange(1, m + 1), n)
    longdf ['row_value'] = occluded_rows.reshape(-1)
    
    E2_group_longdf = E2_group_longdf.append(longdf)
    
E6_files = list(pathlib.Path(path.join('..','experiments','Exp3reference','data','jatos_results_files_batch1')).glob('**/*.json'))

E6_group_longdf = pd.DataFrame()

for i,file in enumerate(E6_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    occluded_rows = np.array(mydf.apply(get_row_vals_three_stimuli, axis=1).to_list())
    
    n, m = occluded_rows.shape
    longdf = pd.DataFrame(np.repeat(mydf.values, m, axis=0), columns=mydf.columns)
    longdf ['row_index'] = np.tile(np.arange(1, m + 1), n)
    longdf ['row_value'] = occluded_rows.reshape(-1)
    
    E6_group_longdf = E6_group_longdf.append(longdf)

E6a_files = list(pathlib.Path(path.join('..','experiments','Exp3referenceLong','data')).glob('**/*.json'))

E6_group_longdf = pd.DataFrame()

for i,file in enumerate(E6_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    occluded_rows = np.array(mydf.apply(get_row_vals_three_stimuli, axis=1).to_list())
    
    n, m = occluded_rows.shape
    longdf = pd.DataFrame(np.repeat(mydf.values, m, axis=0), columns=mydf.columns)
    longdf ['row_index'] = np.tile(np.arange(1, m + 1), n)
    longdf ['row_value'] = occluded_rows.reshape(-1)
    
    E6_group_longdf = E6_group_longdf.append(longdf)
```

```python
# Define a custom function that takes a DataFrame and computes your desired statistic
def compute_stats(group):
    
    if group.loc[group['present']==1,'resp'].mean()==1 or group.loc[group['present']==1,'resp'].mean()==0:
        cond_hit_rate = np.nan
    else:
        cond_hit_rate = group.loc[(group['present'] ==1) & (group['row_value']), 'resp'].mean()
        
    if group.loc[group['present']==0,'resp'].mean()==1 or group.loc[group['present']==0,'resp'].mean()==0:
        cond_fa_rate = np.nan
    else:
        cond_fa_rate = group.loc[(group['present'] ==0) & (group['row_value']), 'resp'].mean()
            
    
    return pd.Series({
        'cond_hit_rate': cond_hit_rate,
        'cond_fa_rate': cond_fa_rate,
        'RT_present': group.loc[(group['present'] ==1) & (group['correct']) & (group['row_value']), 'RT'].median()-
        group.loc[(group['present'] ==1) & (group['correct']), 'RT'].median(),
        'RT_absent': group.loc[(group['present'] ==0) & (group['correct']) & (group['row_value']), 'RT'].median()-
        group.loc[(group['present'] ==0) & (group['correct']), 'RT'].median(),
        'conf_present': group.loc[(group['present'] ==1) & (group['correct']) & (group['row_value']), 'confidence'].mean()-
        group.loc[(group['present'] ==1) & (group['correct']), 'confidence'].mean(),
        'conf_absent': group.loc[(group['present'] ==0) & (group['correct']) & (group['row_value']), 'confidence'].mean()-
        group.loc[(group['present'] ==0) & (group['correct']), 'confidence'].mean()
    })



E4_summary_by_row = E4_group_longdf.groupby(['PROLIFIC_PID','target','row_index']).apply(compute_stats)

def compute_stats_without_confidence(group):
    
    if group.loc[group['present']==1,'resp'].mean()==1 or group.loc[group['present']==1,'resp'].mean()==0:
        cond_hit_rate = np.nan
    else:
        cond_hit_rate = group.loc[(group['present'] ==1) & (group['row_value']), 'resp'].mean()
        
    if group.loc[group['present']==0,'resp'].mean()==1 or group.loc[group['present']==0,'resp'].mean()==0:
        cond_fa_rate = np.nan
    else:
        cond_fa_rate = group.loc[(group['present'] ==0) & (group['row_value']), 'resp'].mean()
            
    
    return pd.Series({
        'cond_hit_rate': cond_hit_rate,
        'cond_fa_rate': cond_fa_rate,
        'RT_present': group.loc[(group['present'] ==1) & (group['correct']) & (group['row_value']), 'RT'].median()-
        group.loc[(group['present'] ==1) & (group['correct']), 'RT'].median(),
        'RT_absent': group.loc[(group['present'] ==0) & (group['correct']) & (group['row_value']), 'RT'].median()-
        group.loc[(group['present'] ==0) & (group['correct']), 'RT'].median()
    })



# E4_summary_by_row = E4_group_longdf.groupby(['PROLIFIC_PID','target','row_index']).apply(compute_stats)
E6_summary_by_row = E6_group_longdf.groupby(['PROLIFIC_PID','target','row_index']).apply(compute_stats_without_confidence)

# E4_group_longdf.groupby('row_index')['row_value'].mean()
```

```python
def compute_time_resolved_kernel(pixel_data, df, target, context):
    mean_yes = np.nanmean(pixel_data[(df['target']==target) & (df['resp']) & (df['context']==context)],axis=0)
    mean_no = np.nanmean(pixel_data[(df['target']==target) & (~df['resp']) & (df['context']==context)],axis=0)
    min_length=min(mean_yes.shape[0],mean_no.shape[0],75)
    time_res_kernel = np.full((75,18,12), np.nan)
    time_res_kernel[:min_length,:,:] = mean_yes[:min_length, :, :]-mean_no[:min_length, :, :]
    return time_res_kernel

individual_time_resolved_kernels = {}
targets = mydf['target'].unique()
contexts = mydf['context'].unique()
responses = [True, False]

for target in targets:
    for context in contexts:
        key = f'{target}_{"cong" if target==context else "incong"}'
        individual_time_resolved_kernels[key] = [];
            

for i,file in enumerate(json_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    max_frames = max(mydf['presented_pixel_data'].apply(lambda x: np.array(x, dtype=float).shape[0]))
    pixel_data_arrays = np.array(mydf.apply(pad_with_nans, axis=1).to_list())
    
    # deman
    mask_present = mydf['present'].values == 1
    mask_target_A = mydf['target'].values == 'A'
    mask_target_S = mydf['target'].values == 'S'

    # Create array of 'max_p' values for broadcasting
    max_p_values = mydf['max_p'].values[:, np.newaxis, np.newaxis, np.newaxis]

    # Use masks and 'max_p' array to update 'pixel_data_arrays'
    pixel_data_arrays[mask_present & mask_target_A] -= max_p_values[mask_present & mask_target_A] * (A-np.mean(A))
    pixel_data_arrays[mask_present & mask_target_S] -= max_p_values[mask_present & mask_target_S] * (S-np.mean(S))

    #done demaning
    
    for target in targets:
        for context in contexts:
            key = f'{target}_{"cong" if target==context else "incong"}'
            individual_time_resolved_kernels[key]+=[compute_time_resolved_kernel(pixel_data_arrays, mydf, target, context)]
    


mean_time_resolved_kernels = {}

for key,value in individual_time_resolved_kernels.items():
    mean_time_resolved_kernels[key] = np.nanmean(np.array(value),axis=0)


```

```python
import imageio
import numpy as np

keys = ['S_cong','S_incong','A_cong','A_incong']


for key in keys:
    
    # Suppose arr is your 3D array
    arr = gaussian_filter(mean_time_resolved_kernels[key],sigma=[1,1,1])  # For example purposes

    # # Normalize array to be in range [0, 255]
    arr = arr-np.min(arr)
    arr = arr/(np.max(arr))
    arr = (arr * 255).astype(np.uint8)

    # Create GIF
    frames = []
    for i in range(arr.shape[0]):
        frames.append(arr[i])

    # Create a black frame
    black_frame = np.zeros((arr.shape[1], arr.shape[2]), dtype=np.uint8)

    # Append black frame to list of frames
    frames.append(black_frame)

    # Save frames as GIF
    frame_rate = 15  # frames per second
    duration = [1.0 / frame_rate] * len(frames)  # duration for each frame
    duration[-1] = 1.0  # duration for the last (black) frame is 1 second
    imageio.mimsave(f'figures/{key}.gif', frames, 'GIF', duration=duration)
```

Plot correlations with A and with S as a function of frame number

```python
def get_time_resolved_correlations(kernel):
    
#     kernel = gaussian_filter(kernel, sigma=2)
    
    n, x, y = kernel.shape
    A_correlations = np.empty(n)
    S_correlations = np.empty(n)

    for i in range(n):
        A_correlation_matrix = np.corrcoef(kernel[i].ravel(), gaussian_filter(A,sigma=1).ravel())
        A_correlations[i] = A_correlation_matrix[0, 1]
        
        S_correlation_matrix = np.corrcoef(kernel[i].ravel(), gaussian_filter(S,sigma=1).ravel())
        S_correlations[i] = S_correlation_matrix[0, 1]
        
    A_df = pd.DataFrame({'reference': 'A', 'frame': range(n), 'correlation': A_correlations})
    S_df = pd.DataFrame({'reference': 'S', 'frame': range(n), 'correlation': S_correlations})
    
    # Concatenate dataframes
    result_df = pd.concat([A_df, S_df], ignore_index=True)

    return result_df
```

```python
df = pd.DataFrame(columns=['target', 'context', 'frame', 'reference', 'correlation'])

for target in targets:
    for context in contexts:
        key = f'{target}_{"cong" if target==context else "incong"}'
        # Assuming get_time_resolved_correlations(kernel) returns a DataFrame
        correlations_df = get_time_resolved_correlations(mean_time_resolved_kernels[key])
        
        # Add 'target' and 'cong' columns
        correlations_df['target'] = target
        correlations_df['context'] = context
        correlations_df['cong'] = "cong" if target==context else "incong"
        
        # Append to the main DataFrame
        df = pd.concat([df, correlations_df], ignore_index=True)
        
df['reference'] = np.where(df['target']==df['reference'], 'target', 'nontarget')
```

```python
A = imageio.imread('stimuli/A.png')[:,:,1]
S = imageio.imread('stimuli/S.png')[:,:,1]

def pad_with_nans(row):
    array = np.array(row['presented_pixel_data'], dtype=float)
    padded_array = np.full((max_frames, array.shape[1], array.shape[2]), np.nan)
    padded_array[:array.shape[0]] = array
    return padded_array

context_dirnames = [path.join('..','experiments','letters5confidence','data','jatos_results_files_batch1'),
                    path.join('..','experiments','letters2','data','jatos_resultfiles_batch1')]  # replace with your directories
json_files = []

for dirname in context_dirnames:
    json_files += list(pathlib.Path(dirname).glob('**/*.json'))

dfs=[]
S_context_kernel = [];
A_context_kernel = [];

pixel_data_array = np.array([])
# json_files = pathlib.Path(dirname).glob('**/*.json')
# first_five_files = list(islice(json_files, 50))
for i,file in enumerate(json_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    max_frames = max(mydf['presented_pixel_data'].apply(lambda x: np.array(x, dtype=float).shape[0]))
    pixel_data_arrays = np.array(mydf.apply(pad_with_nans, axis=1).to_list())
    
    mask_present = mydf['present'].values == 1
    mask_target_A = mydf['target'].values == 'A'
    mask_target_S = mydf['target'].values == 'S'

    # Create array of 'max_p' values for broadcasting
    max_p_values = mydf['max_p'].values[:, np.newaxis, np.newaxis, np.newaxis]

    # Use masks and 'max_p' array to update 'pixel_data_arrays'
    pixel_data_arrays[mask_present & mask_target_A] -= max_p_values[mask_present & mask_target_A] * A
    pixel_data_arrays[mask_present & mask_target_S] -= max_p_values[mask_present & mask_target_S] * S

    subset_yes_S = pixel_data_arrays[(mydf['context']=='S') & (mydf['resp'])]
    subset_no_S = pixel_data_arrays[(mydf['context']=='S') & (~mydf['resp'])]
    dec_kernel_S = np.nanmean(subset_yes_S[:, :, :, :], axis=(0, 1))-np.nanmean(subset_no_S[:, :, :, :], axis=(0, 1))
    S_context_kernel.append(dec_kernel_S)
    
    subset_yes_A = pixel_data_arrays[(mydf['context']=='A') & (mydf['resp'])]
    subset_no_A = pixel_data_arrays[(mydf['context']=='A') & (~mydf['resp'])]
    dec_kernel_A = np.nanmean(subset_yes_A[:, :, :, :], axis=(0, 1))-np.nanmean(subset_no_A[:, :, :, :], axis=(0, 1))
    A_context_kernel.append(dec_kernel_A)
    
```

```python


targets = {'A': A, 'S':S}

def get_correlation_with_S(row):
    if str(row['presented_pixel_data'])=='nan':
        return 0
    else:
        presented_pixel_data = np.array(row['presented_pixel_data'], dtype=float)
        correlation_per_frame = np.array([ma.corrcoef(ma.masked_invalid(frame.flatten()),
                                            ma.masked_invalid(S.flatten()))[0,1] for frame in presented_pixel_data])
        return correlation_per_frame
    
def get_correlation_with_A(row):
    if str(row['presented_pixel_data'])=='nan':
        return 0
    else:
        presented_pixel_data = np.array(row['presented_pixel_data'], dtype=float)
        correlation_per_frame = np.array([ma.corrcoef(ma.masked_invalid(frame.flatten()),
                                            ma.masked_invalid(A.flatten()))[0,1] for frame in presented_pixel_data])
        return correlation_per_frame
    
def get_correlation_with_target_letter(row):
    if row.target=='S':
        return get_correlation_with_S(row)
    elif row.target=='A':
        return(get_correlation_with_A(row))
    
def get_frame_data(row):
    if str(row['presented_pixel_data'])=='nan':
        return 0
    else:
        presented_pixel_data = np.array(row['presented_pixel_data'], dtype=float)
        num_frames = presented.pixel.data.shape[0]
        flattened_strings = []
        for frame in range(num_frames):
            flattened_array = presented_pixel_data[frame].flatten()
            flattened_string = ','.join(map(str, flattened_array))
            flattened_strings.append(flattened_string)
        return flattened_strings
    
def get_correlation_with_mask(row):
    if str(row['presented_pixel_data'])=='nan':
        return 0
    else:
        presented_pixel_data = np.array(row['presented_pixel_data'], dtype=float)
        target = targets[row['target']];
        correlation_with_nans = ma.corrcoef(np.where(np.isnan(presented_pixel_data[0,:,:].flatten()),1,0),
                                            target.flatten())[0,1]
        return correlation_with_nans
```

```python
def get_num_pixels_hidden(row):
    if str(row['presented_pixel_data'])=='nan':
        return 0
    else:
        presented_pixel_data = row['presented_pixel_data']
        first_frame = presented_pixel_data[0];
        number_of_hidden_pixels = len([item for sublist in first_frame for item in sublist if str(item) == 'None'])
        return number_of_hidden_pixels

def to_csv_multiple_files(dirname):
    dfs=[]
    json_files = glob.glob(path.join(dirname,'*','*.json'))
    for i,file in enumerate(json_files):
                mydf = pd.read_json(file)
                mydf['num_pixels_hidden'] = mydf.apply(get_num_pixels_hidden, axis=1)
#                 mydf['correlation_with_A'] = mydf.apply(get_correlation_with_A, axis=1)   
#                 mydf['correlation_with_S'] = mydf.apply(get_correlation_with_S, axis=1) 
                mydf['correlation_with_target_letter'] = mydf.apply(get_correlation_with_target_letter, axis=1) 
#                 mydf['frame_data'] = mydf.apply(get_frame_data,axis=1)
                mydf['correlation_with_mask'] = mydf.apply(get_correlation_with_mask, axis=1)
                mydf = mydf.explode('correlation_with_target_letter');
                mydf['frame_index'] = mydf.groupby('trial_index').cumcount()
                dfs.append(mydf.drop('presented_pixel_data',axis=1));
    group_df = pd.concat(dfs)
    group_df.to_csv(dirname+'/all_data.csv', index=False)
    return(group_df)

#12.03.2023: JATOS changed their export data format (the tree structure is more convolved now)
def to_csv_multiple_files_new_format(dirname):
    dfs=[]
    json_files = pathlib.Path(dirname).glob('**/*.json')
    for i,file in enumerate(json_files):
                mydf = pd.read_json(file)
                mydf['num_pixels_hidden'] = mydf.apply(get_num_pixels_hidden, axis=1)
                mydf['correlation_with_target_letter'] = mydf.apply(get_correlation_with_target_letter, axis=1)   
                mydf['correlation_with_mask'] = mydf.apply(get_correlation_with_mask, axis=1)
                mydf = mydf.explode('correlation_with_target_letter');
                mydf['frame_index'] = mydf.groupby('trial_index').cumcount()
                dfs.append(mydf.drop('presented_pixel_data',axis=1));
    group_df = pd.concat(dfs)
    group_df.to_csv(dirname+'/all_data.csv', index=False)
    return(group_df)
```
