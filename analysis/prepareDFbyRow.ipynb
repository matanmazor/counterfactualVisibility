Extract row occlusion decision kernels

```python
def get_row_vals(row):
    array = np.array(row['presented_pixel_data'], dtype=float)
    mean_array = np.mean(array,axis=(0,2))
    return np.isnan(mean_array)

def get_row_vals_three_stimuli(row):
    array = np.array(row['presented_pixel_data'], dtype=float)
    mean_array = np.mean(array,axis=(0,1,3))
    return np.isnan(mean_array)

E2_files = list(pathlib.Path(path.join('..','experiments','Exp2rows','data','jatos_resultfiles_batch1')).glob('**/*.json'))

E2_group_longdf = pd.DataFrame()

for i,file in enumerate(E2_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    occluded_rows = np.array(mydf.apply(get_row_vals, axis=1).to_list())
    
    n, m = occluded_rows.shape
    longdf = pd.DataFrame(np.repeat(mydf.values, m, axis=0), columns=mydf.columns)
    longdf ['row_index'] = np.tile(np.arange(1, m + 1), n)
    longdf ['row_value'] = occluded_rows.reshape(-1)
    
    E2_group_longdf = E2_group_longdf.append(longdf)
    
E3_files = list(pathlib.Path(path.join('..','experiments','Exp3reference','data','jatos_results_files_batch1')).glob('**/*.json'))

E3_group_longdf = pd.DataFrame()

for i,file in enumerate(E3_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    occluded_rows = np.array(mydf.apply(get_row_vals_three_stimuli, axis=1).to_list())
    
    n, m = occluded_rows.shape
    longdf = pd.DataFrame(np.repeat(mydf.values, m, axis=0), columns=mydf.columns)
    longdf ['row_index'] = np.tile(np.arange(1, m + 1), n)
    longdf ['row_value'] = occluded_rows.reshape(-1)
    
    E3_group_longdf = E3_group_longdf.append(longdf)

E6a_files = list(pathlib.Path(path.join('..','experiments','Exp3referenceLong','data')).glob('**/*.json'))

E3_group_longdf = pd.DataFrame()

for i,file in enumerate(E3_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    occluded_rows = np.array(mydf.apply(get_row_vals_three_stimuli, axis=1).to_list())
    
    n, m = occluded_rows.shape
    longdf = pd.DataFrame(np.repeat(mydf.values, m, axis=0), columns=mydf.columns)
    longdf ['row_index'] = np.tile(np.arange(1, m + 1), n)
    longdf ['row_value'] = occluded_rows.reshape(-1)
    
    E3_group_longdf = E3_group_longdf.append(longdf)
```

```python
# Define a custom function that takes a DataFrame and computes your desired statistic
def compute_stats(group):
    
    if group.loc[group['present']==1,'resp'].mean()==1 or group.loc[group['present']==1,'resp'].mean()==0:
        cond_hit_rate = np.nan
    else:
        cond_hit_rate = group.loc[(group['present'] ==1) & (group['row_value']), 'resp'].mean()
        
    if group.loc[group['present']==0,'resp'].mean()==1 or group.loc[group['present']==0,'resp'].mean()==0:
        cond_fa_rate = np.nan
    else:
        cond_fa_rate = group.loc[(group['present'] ==0) & (group['row_value']), 'resp'].mean()
            
    
    return pd.Series({
        'cond_hit_rate': cond_hit_rate,
        'cond_fa_rate': cond_fa_rate,
        'RT_present': group.loc[(group['present'] ==1) & (group['correct']) & (group['row_value']), 'RT'].median()-
        group.loc[(group['present'] ==1) & (group['correct']), 'RT'].median(),
        'RT_absent': group.loc[(group['present'] ==0) & (group['correct']) & (group['row_value']), 'RT'].median()-
        group.loc[(group['present'] ==0) & (group['correct']), 'RT'].median(),
        'conf_present': group.loc[(group['present'] ==1) & (group['correct']) & (group['row_value']), 'confidence'].mean()-
        group.loc[(group['present'] ==1) & (group['correct']), 'confidence'].mean(),
        'conf_absent': group.loc[(group['present'] ==0) & (group['correct']) & (group['row_value']), 'confidence'].mean()-
        group.loc[(group['present'] ==0) & (group['correct']), 'confidence'].mean()
    })



E2_summary_by_row = E2_group_longdf.groupby(['PROLIFIC_PID','target','row_index']).apply(compute_stats)

def compute_stats_without_confidence(group):
    
    if group.loc[group['present']==1,'resp'].mean()==1 or group.loc[group['present']==1,'resp'].mean()==0:
        cond_hit_rate = np.nan
    else:
        cond_hit_rate = group.loc[(group['present'] ==1) & (group['row_value']), 'resp'].mean()
        
    if group.loc[group['present']==0,'resp'].mean()==1 or group.loc[group['present']==0,'resp'].mean()==0:
        cond_fa_rate = np.nan
    else:
        cond_fa_rate = group.loc[(group['present'] ==0) & (group['row_value']), 'resp'].mean()
            
    
    return pd.Series({
        'cond_hit_rate': cond_hit_rate,
        'cond_fa_rate': cond_fa_rate,
        'RT_present': group.loc[(group['present'] ==1) & (group['correct']) & (group['row_value']), 'RT'].median()-
        group.loc[(group['present'] ==1) & (group['correct']), 'RT'].median(),
        'RT_absent': group.loc[(group['present'] ==0) & (group['correct']) & (group['row_value']), 'RT'].median()-
        group.loc[(group['present'] ==0) & (group['correct']), 'RT'].median()
    })



E3_summary_by_row = E3_group_longdf.groupby(['PROLIFIC_PID','target','row_index']).apply(compute_stats_without_confidence)

# E4_group_longdf.groupby('row_index')['row_value'].mean()
```
