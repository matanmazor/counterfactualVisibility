```python
import json
import pandas as pd
import numpy as np
from os import path as path
import glob
import imageio
import numpy.ma as ma
import pathlib
from itertools import islice
import scipy

def to_csv(filename):
    dfs=[]
    with open(filename+'.txt') as json_file:
        for i,line in enumerate(json_file):
            if line[0:14]!='Consent given.':
                dfs.append(pd.read_json(line))
    group_df = pd.concat(dfs)
    group_df.to_csv(filename+'.csv', index=False)
    return(group_df)

# group_df = to_csv(path.join('..','experiments','pilots','letters','data','jatos_results_batch3'))
```

Extract row occlusion decision kernels

```python
def get_row_vals(row):
    array = np.array(row['presented_pixel_data'], dtype=float)
    mean_array = np.mean(array,axis=(0,2))
    return np.isnan(mean_array)

def get_row_vals_three_stimuli(row):
    array = np.array(row['presented_pixel_data'], dtype=float)
    mean_array = np.mean(array,axis=(0,1,3))
    return np.isnan(mean_array)

def compute_decision_kernel(occluded_rows, df, target):
    subset_yes = occluded_rows[(df['target']==target) & (df['resp'])]
    subset_no = occluded_rows[(df['target']==target) & (~df['resp'])]
    return np.nanmean(subset_yes, axis=0)-np.nanmean(subset_no, axis=0)

def compute_hit_kernel(occluded_rows, df, target):
    hit_rate = df[(df['target']==target) & (df['present']==1)]['resp'].mean()
    subset_yes = occluded_rows[(df['target']==target) & (df['present']==1) & (df['resp'])]
    p_occluded = occluded_rows[(df['target']==target) & (df['present']==1)]
    #Bayes rule: p(yes|occluded)=p(occluded|yes)p(yes)/p(occluded)
    return hit_rate*np.nanmean(subset_yes, axis=0)/np.nanmean(p_occluded, axis=0)

def compute_fa_kernel(occluded_rows, df, target):
    fa_rate = df[(df['target']==target) & (df['present']==0)]['resp'].mean()
    subset_yes = occluded_rows[(df['target']==target) & (df['present']==0) & (df['resp'])]
    p_occluded = occluded_rows[(df['target']==target) & (df['present']==0)]
    #Bayes rule: p(yes|occluded)=p(occluded|yes)p(yes)/p(occluded)
    return fa_rate*np.nanmean(subset_yes, axis=0)/np.nanmean(p_occluded, axis=0)

def compute_RT_kernel(occluded_rows, df, target, present):
    present = ['absent','present'].index(present)
    median_RT = df[(df['target']==target) & (df['correct']) & (df['present']==present)]['RT'].median()
    subset_slow = occluded_rows[(df['target']==target) & (df['correct']) & (df['present']==present) & (df['RT']>=median_RT)]
    subset_fast = occluded_rows[(df['target']==target) & (df['correct']) & (df['present']==present) & (df['RT']<median_RT)]
    return np.nanmean(subset_slow, axis=0)-np.nanmean(subset_fast, axis=0)

def compute_confidence_kernel(occluded_rows, df, target, present):
    present = ['absent','present'].index(present)
    mean_confidence = df[(df['target']==target) & (df['correct']) & (df['present']==present)]['confidence'].mean()
    subset_high = occluded_rows[(df['target']==target) & (df['correct']) & (df['present']==present) & (df['confidence']>=mean_confidence)]
    subset_low = occluded_rows[(df['target']==target) & (df['correct']) & (df['present']==present) & (df['confidence']<mean_confidence)]
    return np.nanmean(subset_high, axis=0)-np.nanmean(subset_low, axis=0)

# Define a function to compute median RT where correct == True
def compute_median(group):
    return group[(group['correct'] == True) and group['test_part'] in ['test1','test2']]['RT'].median()

E2_files = list(pathlib.Path(path.join('..','experiments','Exp2rows','data','jatos_resultfiles_batch1')).glob('**/*.json'))

individual_E2_kernels = {}
individual_E3_kernels = {}

targets = ['S','A']#pd.read_json(E2_files[0])['target'].unique()
responses = [True, False]

for target in targets:
    for individual_kernels in [individual_E2_kernels, individual_E3_kernels]:
        individual_kernels[target] = [];
        individual_kernels[f'{target}_RT_present'] = [];
        individual_kernels[f'{target}_RT_absent'] = [];
        individual_kernels[f'{target}_conf_present'] = [];
        individual_kernels[f'{target}_conf_absent'] = [];
        individual_kernels[f'{target}_hit'] = [];
        individual_kernels[f'{target}_fa'] = [];

for i,file in enumerate(E2_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    occluded_rows = np.array(mydf.apply(get_row_vals, axis=1).to_list())
    
    for target in targets:
        key = target
        individual_E2_kernels[key]+=[compute_decision_kernel(occluded_rows, mydf, target)]
        individual_E2_kernels[f'{key}_hit']+=[compute_hit_kernel(occluded_rows, mydf, target)]
        individual_E2_kernels[f'{key}_fa']+=[compute_fa_kernel(occluded_rows, mydf, target)]
        
        for present in ['present','absent']:
            key=f'{target}_RT_{present}'
            individual_E2_kernels[key]+=[compute_RT_kernel(occluded_rows, mydf, target,present)]
            key=f'{target}_conf_{present}'
            individual_E2_kernels[key]+=[compute_confidence_kernel(occluded_rows, mydf, target,present)]
    
E3_files = list(pathlib.Path(path.join('..','experiments','Exp3reference','data','jatos_results_files_batch1')).glob('**/*.json'))

for i,file in enumerate(E3_files):
    mydf = pd.read_json(file)
    mydf = mydf[mydf.test_part.isin(['test1','test2'])]
    mydf['resp'] = mydf.response==mydf.presence_key
    max_frames = max(mydf['presented_pixel_data'].apply(lambda x: np.array(x, dtype=float).shape[0]))
    occluded_rows = np.array(mydf.apply(get_row_vals_three_stimuli, axis=1).to_list())
    
    for target in targets:
        key = target
        individual_E3_kernels[key]+=[compute_decision_kernel(occluded_rows, mydf, target)]
        individual_E3_kernels[f'{key}_hit']+=[compute_hit_kernel(occluded_rows, mydf, target)]
        individual_E3_kernels[f'{key}_fa']+=[compute_fa_kernel(occluded_rows, mydf, target)]

        for present in ['present','absent']:
            key=f'{target}_RT_{present}'
            individual_E3_kernels[key]+=[compute_RT_kernel(occluded_rows, mydf, target,present)]


mean_E2_kernels = {}

for key,value in individual_E2_kernels.items():
    mean_E2_kernels[key] = np.nanmean(np.array(value),axis=0)


t_E2_kernels = {}

for key,value in individual_E2_kernels.items():
    t_E2_kernels[key] = scipy.stats.ttest_1samp(np.array(pd.DataFrame(value).dropna()),0,axis=0).statistic
    
mean_E3_kernels = {}

for key,value in individual_E3_kernels.items():
    mean_E3_kernels[key] = np.nanmean(np.array(value),axis=0)


t_E3_kernels = {}

for key,value in individual_E3_kernels.items():
    t_E3_kernels[key] = scipy.stats.ttest_1samp(np.array(pd.DataFrame(value).dropna()),0,axis=0).statistic
```
